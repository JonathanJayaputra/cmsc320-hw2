```python
# Not a jerk = 0, Mildly a jerk = 1, Strongly a jerk = 2
aita_values = [aita_questions[i] + " numeric" for i in range(len(aita_questions))] 
aita_values
```

```python
for i in range(len(aita_questions)):
    df1[aita_values[i]] = df1[aita_questions[i]].map({'Not a jerk': 0, 'Mildly a jerk': 1, 'Strongly a jerk': 2})
    df2[aita_values[i]] = df2[aita_questions[i]].map({'Not a jerk': 0, 'Mildly a jerk': 1, 'Strongly a jerk': 2})
    df3[aita_values[i]] = df3[aita_questions[i]].map({'Not a jerk': 0, 'Mildly a jerk': 1, 'Strongly a jerk': 2})
    df4[aita_values[i]] = df4[aita_questions[i]].map({'Not a jerk': 0, 'Mildly a jerk': 1, 'Strongly a jerk': 2})
```

```python
data = {'upbringing': df['upbringing'], 'reading': df['reading numeric']}
df_bubble = pd.DataFrame(data)

# Count occurrences of each (upbringing, reading) pair
bubble_data = df_bubble.groupby(['upbringing', 'reading']).size().reset_index(name='count')

plt.figure(figsize=(8, 6))

# Create scatter plot where size represents frequency
bubble_plot = sns.scatterplot(
    x='upbringing', 
    y='reading', 
    size='count', 
    sizes=(50, 500),  # Adjust size range
    data=bubble_data,
    alpha=0.6, 
    edgecolor='black'
)

# Improve visibility
plt.title("Bubble Plot of Reading Scores by Upbringing")
plt.yticks([0, 1, 2])  # Ensure y-axis shows categorical values
plt.grid(axis='y', linestyle='--', alpha=0.7)
bubble_plot.legend_.remove()

plt.show()
```
