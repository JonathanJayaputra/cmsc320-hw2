```python
#drops dont know its complicated and strongly conservative

def graph_proportion(name, demo):
    political_order = [
        "Strongly liberal",
        "Mildly liberal",
        "Neutral",
        "Don't know / It's complicated",
        "Mildly conservative",
        "Strongly conservative"
    ]

    political_order = [
        "Strongly liberal",
        "Mildly liberal",
        "Neutral",
        "Mildly conservative",
    ]

    # Assign corresponding colors
    political_colors = {
        "Strongly liberal": "#0000cc",  # Deep Blue
        "Mildly liberal": "#3366ff",    # Medium Blue
        "Neutral": "#666666",           # Dark Gray
        "Don't know / It's complicated": "#ffcc00",  # Yellow
        "Mildly conservative": "#ff5050",  # Medium Red
        "Strongly conservative": "#cc0000"  # Deep Red
    }

    political_colors = {
    "Strongly liberal": "#0000cc",  # Deep Blue
    "Mildly liberal": "#3366ff",    # Medium Blue
    "Neutral": "#b9abc8",           # Dark Gray
    "Mildly conservative": "#ff5050",  # Medium Red
    "Strongly conservative": "#cc0000",  # Deep Red
    }

    df_political = df_all[df_all[demo].isin(political_order)]
    # Step 1: Count responses by answer (doctor) and political leaning (self)
    group_counts = df_political.groupby([name, demo]).size().unstack(fill_value=0)

    # Step 2: Normalize each political group to the same size
    min_group_size = group_counts.sum(axis=0).min()
    normalized_counts = (group_counts.div(group_counts.sum(axis=0), axis=1) * min_group_size).astype(int)

    # Step 3: Convert to proportions
    normalized_proportions = normalized_counts.div(normalized_counts.sum(axis=1), axis=0)

    # Step 4: Reorder columns based on political spectrum
    normalized_proportions = normalized_proportions[political_order]

    jerk_labels = {0: "Not a jerk", 1: "Mildly a jerk", 2: "Strongly a jerk"}
    jerk_order = ["Not a jerk", "Mildly a jerk", "Strongly a jerk"]

    normalized_proportions = normalized_proportions.reindex(jerk_order)

    # Step 5: Plot the stacked bar chart with sorted colors
    ax = normalized_proportions.plot(
        kind='bar', stacked=True, color=[political_colors[cat] for cat in political_order], figsize=(8, 6)
    )



    plt.ylabel("Proportion")
    plt.xlabel("Jerk Rating " + name)
    plt.title(f"Political Leaning Distribution for {name} Jerk Rating")
    plt.legend(title="Political Leaning", labels=political_order, loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(rotation=0)
    plt.show()

for value in aita_questions:
    graph_proportion(value, "self")
```
